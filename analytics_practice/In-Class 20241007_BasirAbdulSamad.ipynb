```python
# exploring the dataset in a similar way as in-class but this time using pivot tables
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_df = pd.read_csv(url)
titanic_df.head()
```

```python
# my question 1: how do survival rates vary by class and gender?
survival_rate_pivot = titanic_df.pivot_table(values = 'Survived', index = 'Pclass', columns = 'Sex', aggfunc = 'mean').round(2)
survival_rate_pivot
```

```python
# my question 2: how has age impacted survival?
age_pivot = titanic_df.pivot_table(values = 'Age', index = 'Survived', columns = 'Pclass', aggfunc = 'mean').round(2)
age_pivot
```

```python
# my question 3: can we find a significant difference in the fares paid by survivors and non-survivors?
fare_pivot = titanic_df.pivot_table(values = 'Fare', index = 'Survived', columns = 'Pclass', aggfunc = 'mean').round(2)
fare_pivot
```

```python
# my question 4: how did the average fare vary across different class between males and females?
avgFare_pivot = titanic_df.pivot_table(values = 'Fare', index = 'Pclass', columns = 'Sex', aggfunc = 'mean').round(2)
avgFare_pivot
```

```python
# my question 5: what could have been the survival rate of the elderly (more than 64 years old) by sex?
titanic_df['Old'] = titanic_df['Age'] > 64 # a new column to specify passengers as the elderly

#back to pivotting
child_survival_pivot = titanic_df.pivot_table(values = 'Survived', index = 'Old', columns = 'Sex', aggfunc = 'mean').round(2)
child_survival_pivot
```

Using pivot tables was fun where switching around and playing with the index, values, and column parameters made completing this assignment much easier. I could make questions based on the parameters I came up with and to go over my analysis which revealed some signficant insights. Those were insights mainly focused on the survival rates on board, as only the 4th question checked for the average amount of fares paid by different genders. To start off with the survival rates in the first two questions, it showed that the survival rates varied by a lot by both gender and class, especially 3rd class where the survival rate was dramtically lower than the rest. From questions 3 and 4, hat was signficant about the fares was that the average fare accounted for first-class passengers was notably higher than that for lower classes, reflecting the socioeconomic disparity. Last question explored about the survival rate in the elderly (which happened to be not very meaningful) using through the "Old" column, where it was discovered that there were barely any old people on board, while the male percentage of old people survival rate was only 9%, whereas none for females older than 64.
