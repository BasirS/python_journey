```python
# pandas II - assignment 1 re-do
import pandas as pd
import datetime as dt
import calendar

# approach 1: using a set (only unique values)
def create_full_calendar(year, month):
    eq1 = pd.read_csv("earthquakes.csv")
    eq2 = pd.read_csv("./earthquakes_2023_global.csv")
    
    eq_dates = []
    
    for date in eq1['time']:
        date = dt.datetime.fromtimestamp(date/1000)
        date = date.replace(hour=0, minute=0, second=0, microsecond=0)
        eq_dates.append(date)
    
    for date in eq2['time']:
        date = dt.datetime.fromisoformat(date)
        date = date.replace(hour=0, minute=0, second=0, microsecond=0, tzinfo=None)
        eq_dates.append(date)
    
    eq_dates = set(eq_dates)
    
    first_weekday, days_in_month = calendar.monthrange(year, month)
    
    day_counter = 1
    output_str = f"    {calendar.month_name[month]} {year}\n"
    output_str += "Su Mo Tu We Th Fr Sa\n"
    
    # Add initial spacing for the first week
    output_str += "   " * first_weekday
    
    while day_counter <= days_in_month:
        for weekday in range(7):
            if day_counter > days_in_month:
                break  # Break the loop if the last day is reached
            
            # Print the day number
            if dt.datetime(year, month, day_counter) in eq_dates:
                output_str += f"{day_counter:2d} "
            else:
                output_str += f"{' X '}"
            
            # Check if we need to move to the next line
            if (first_weekday + day_counter) % 7 == 0:
                output_str += "\n"
            
            day_counter += 1
    
    return output_str
```

```python
import numpy as np

# approach 2: using ^^
def create_full_calendar(year, month):
    eq1 = pd.read_csv("./earthquakes.csv")
    eq2 = pd.read_csv("./earthquakes_2023_global.csv")
    

    eq1['time'] = pd.to_datetime(eq1['time'], unit='ms')
    eq2['time'] = pd.to_datetime(eq2['time'])

    dates = np.concatenate([eq1['time'].dt.date, eq2['time'].dt.date])

    first_weekday, days_in_month = calendar.monthrange(year, month)
    
    day_counter = 1
    output_str = f"    {calendar.month_name[month]} {year}\n"
    output_str += "Su Mo Tu We Th Fr Sa\n"
    
    # Add initial spacing for the first week
    output_str += "   " * first_weekday
    
    while day_counter <= days_in_month:
        for weekday in range(7):
            if day_counter > days_in_month:
                break  # Break the loop if the last day is reached
            
            # Print the day number
            if dt.date(year, month, day_counter) in dates:
                output_str += f"{day_counter:2d} "
            else:
                output_str += f"{' X '}"
            
            # Check if we need to move to the next line
            if (first_weekday + day_counter) % 7 == 0:
                output_str += "\n"
            
            day_counter += 1
    
    return output_str

print(create_full_calendar(2024,1))
print(create_full_calendar(2023,1))
print(create_full_calendar(2018,9))
print(create_full_calendar(2018,10))
```

```python
null_embarked = titanic_df[(titanic_df['Age'].isnull())]
print(null_embarked)
```

```python
# fill missing Age values with the median age
titanic_cleaned = titanic_df
titanic_cleaned['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())
```

```python
# drop rows where 'Embarked' is missing
titanic_cleaned = titanic_cleaned.dropna(subset=['Embarked'])

titanic_cleaned.isnull().sum()
```

```python
# basic exploration
# how many people survived?
survival_counts = titanic_df['Survived'].value_counts()
survival_counts
```

```python
average_age_survival = titanic_df.groupby('Survived')['Age'].mean().round()
average_age_survival
# average age of survivors vs non-survivors
```

```python
average_fare_class = titanic_df.groupby('Pclass')['Fare'].mean().round()
average_fare_class
# average fare paid by class
```

```python
young_survivors = titanic_df[(titanic_df['Age'] < 18) & (titanic_df['Survived'] == 1)]
young_survivors[['Name', 'Age', 'Sex', 'Survived']].head()
```

```python
# filter passengers in 1st class who paid more than 100 fare
ticket_price = titanic_cleaned[(titanic_cleaned['Pclass'] == 1) & (titanic_cleaned['Fare'] > 100)]
ticket_price
```

```python
# group by gender and calculate survival rate
gender_survival_rate = titanic_df.groupby('Sex')['Survived'].mean().round(2)
gender_survival_rate
```
