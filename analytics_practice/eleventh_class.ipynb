```python
# data viz 
%pip install seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb


#read the diabetes dataframe
df = pd. read_csv('diabetes.csv')
df
```

```python
#a really nice looking correlation matrix
corrMatrix = df.corr()
plt.figure(figsize=(10,6))
sb.heatmap(corrMatrix)
```

```python
#diagram of what strong correlation looks like
# dfgr = df.groupby(['Age'],as_index=False)['Pregnancies'].sum()
dfgr = df.groupby(['Age'],as_index=False)['Pregnancies'].mean()
plt.style.use('dark_background')
plt.scatter(dfgr['Age'],dfgr["Pregnancies"])
plt.xlabel("Age")
plt.ylabel("Pregnancies")
```

```python
#just a quick diagram of what weak correlation looks like
dfgr = df.groupby(['SkinThickness'],as_index=False)['Glucose'].sum()
plt.scatter(dfgr["SkinThickness"],dfgr["Glucose"])
plt.xlabel("Skin Thickness")
plt.ylabel("Glucose")
```
