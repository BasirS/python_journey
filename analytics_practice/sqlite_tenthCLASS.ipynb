```python
# today's agenda: the ITIS database can be accessed/loaded locally

import sqlite3
import pandas as pd

csv_file = 'comptab_2018-01-29 16_00_comma_separated.csv'
df_csv = pd.read_csv(csv_file)
df_csv.tail()
```

```python
print(df_csv['Term'].unique())
```

```python
db_path = 'ITIS.sqlite'
conn = sqlite3.connect(db_path)

query = "SELECT tsn, completename FROM longnames"

df_db = pd.read_sql_query(query, conn)

conn.close()

df_db
```

```python
df_merged = pd.merge(df_csv, df_db, left_on = 'Taxon', right_on = 'completename')
df_merged
```

```python
conn = sqlite3.connect(db_path)

query = "SELECT vernacular_name, tsn FROM vernaculars WHERE language = 'English'"

df_vernaculars = pd.read_sql_query(query, conn)

conn.close()

df_vernaculars
```

```python
df_with_human_names = pd.merge(df_merged, df_vernaculars, on = "tsn")

df_with_human_names
```

```python
# "straight-up pandas" from here on out
df_endangered = df_with_human_names[(df_with_human_names['App.'] == "I")]

df_endangered
```

```python
print(df_endangered[df_endangered['Term'] == 'jewellery - ivory ']['vernacular_name'].unique())
```

```python
# in-class insights
db_path = 'ITIS.sqlite'
conn = sqlite3.connect(db_path)

query2 = "SELECT tsn, geographic_value FROM geographic_div"

df_db2 = pd.read_sql_query(query2, conn)

conn.close()

df_db2 
```

```python
df_final = pd.merge(df_merged, df_db2, on = 'tsn', how = 'left')
df_final
```

```python
# exploring trends by comparing importers with exporters (csv file) through the calculations of discrepencies between the two
df_with_human_names['quantity_discrepency'] = df_with_human_names['Importer reported quantity'] - df_with_human_names['Exporter reported quantity']

# creating a dataframe that includes rows where qualuty_discrepency includes values that are not empty (NaN)
discrepency_summary = df_with_human_names[df_with_human_names['quantity_discrepency'].notna()]

# to output large discrepencies
discrepency_summary.sort_values(by='quantity_discrepency', ascending = False)
```

```python
print(discrepency_summary['quantity_discrepency'].describe().round(1))
```

A lot of the originally occuring NaN values in "Importer reported quantity" and "Exporter reported quantity" columns are no longer present after adding a new column called "quality_discrepency", which dropped all the NaN values because of the second line of code. We can see that the first five rows (head of the table) shows positive values, which depicts that our third line of code worked for outputting values in descending order, indicating that the importer reported receiving more than the exporter reported sending. Then, we also see negative values for the last five rows (tail of the table), where it means that the importer reported receiving less than the exporter reported sending. To gain more insight, having the overall statistical summary of the quality_discrepency column gives us an idea on how the discrepency averages around -816.5 and returns minimum and maximum discrepancy values of -2768688.0 and 220000.0, respectively. The summary signifies that there was strong variability in reporting between importers and exporters.
