```python
# let's import the dependencies
import numpy as np
import pandas as pd

data = pd.read_csv("diagnostic_data.csv")
data
```

```python
# checking the different features present in the dataset
data.head()
```

```python
# checking for its shape - diagnosis column is our TARGET
data.shape
```

```python
# checking the data type of each columns and non-null count
data.info()
```

```python
# time to check for the thing i dislike the most in my datasets: empty/missing values!
print(data.isna().sum())
```

```python
# fixing the index and removing the last column for a cleaner dataset
data = data.set_index('id')
data = data.drop(columns=['Unnamed: 32'])
data
```

```python
# with 'id' and 'unnamed: 32' out of the way, let's get some stats
data.describe()
```

```python
# understanding the diagnosis column by selecting M and B to compare stats between the two 
malignant_cases = data.loc[data['diagnosis'] == 'M']
benign_cases = data.loc[data['diagnosis'] == 'B']
malignant_mean = malignant_cases.mean(numeric_only=True)
benign_mean = benign_cases.mean(numeric_only=True)
print(malignant_mean)
print(benign_mean)
```

```python
# checking out rows 10 to 25 and columns 1 to 5
data.iloc[10:25, 1:6]
```

```python
# ignoring every other column and focusing on learning about the first 30 rows to know what IDs have the diagnosis as benign or malignant
print(data.iloc[:30, :1])
```

```python
# learning about the two diagnosis groups regarding their average feature values
data.groupby('diagnosis').mean()
```

```python
# now, checking for the largest area of the tumor cells, since a larger tumor could be factor in malignancy
max_area_worst = data['area_worst'].max()
max_area_worst
```

```python
# lastly, sorting the diagnostic dataset with area_worst to see the first 5 cases that appear for occupying the largest area
data_sorted = data.sort_values(by = 'area_worst', ascending = False)
data_sorted.head()
```
