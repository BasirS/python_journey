```python
temps['datetime'] = pd.to_datetime(temps['datetime'])
temps.head()
```

```python
temps.set_index('datetime', inplace = True)
```

```python
temps
```

```python
temps.columns
```

```python
# what was the average temperature in Vancouver over these observations

print(temps['Vancouver'].mean())
```

```python
# 283.9 is very hot but the thing is that it is in Kelvins, so now we try to convert it to F degrees
# tF = (tK - 273.15) * (9/5) + 32

temps = (temps - 273.15) * (9/5) + 32
temps
```

```python
temps.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
temps.plot(figsize = (10,6), title = 'Temperatures Over Time (Fahrenheit)', grid = True)

plt.xlabel('Date')
plt.ylabel('Temperature (°F)')

plt.show()

#What do I want to do with this dataset? My thought process; what might I want to learn, observe, and predict from this?

```

```python
# REST API BASICS

## What is an API?
# An API (Application Programming Interface) is a set of rules that allows two applications to communicate with each other. 
# It acts as a bridge for exchanging data between a client (like a web browser) and a server.

## API Communication Types
# - **GET**: Requests data from the server (READ operation).
# - **POST**: Sends data to the server (WRITE operation).
# 
# Example:
# - **GET /resource**: Fetches a resource from the server.
# - **POST /resource**: Submits new data to the server.

## HTTP Status Codes
# When a request is made to the server, the server responds with a status code indicating the success or failure of the request.
# Common status codes:
# - **200 OK**: The request was successful, and the server returned the requested data.
# - **404 NOT FOUND**: The requested resource could not be found, and this is the most common type of error.

# Example of a process flow:
# 1. The client requests a webpage (GET /page).
# 2. The server responds with **200 OK** and the webpage content (HTML).
# 3. The user clicks a link, making another GET request (GET /another-resource).
# 4. The server responds with **200 OK** again.

## Traditional Web Applications
# Traditional web applications rely on HTML and HTTP protocols to build their user interfaces.
# However, this approach is not always optimal for modern applications, especially those on mobile devices.
# Mobile applications (iOS, Android) use GUI (Graphical User Interface) built with different languages (e.g., Swift for iOS, Kotlin for Android) rather than HTML.

## API Types
# There are different types of APIs used in web applications, including:
# - **RPC (Remote Procedure Call)**: Allows execution of a procedure on a remote server.
# - **RMI (Remote Method Invocation)**: A Java-specific API that allows objects to invoke methods on an object running on another
```
